```python
import sympy as sp

theta1, theta2, theta3, theta4, theta5, theta6 = sp.symbols('theta1 theta2 theta3 theta4 theta5 theta6')
d2, d6 = sp.symbols('d2 d6')

c1, s1 = sp.cos(theta1), sp.sin(theta1)
c2, s2 = sp.cos(theta2), sp.sin(theta2)
c3p = sp.cos(sp.pi/2 + theta3)
s3p = sp.sin(sp.pi/2 + theta3)
c4, s4 = sp.cos(theta4), sp.sin(theta4)
c5m = sp.cos(sp.pi/2 - theta5)
s5m = sp.sin(sp.pi/2 - theta5)
c6, s6 = sp.cos(theta6), sp.sin(theta6)

# A1
A1 = sp.Matrix([
    [c1, 0,  s1, 0],
    [s1, 0, -c1, 0],
    [0,  1,  0,  13],
    [0,  0,  0,  1]
])

# A2
A2 = sp.Matrix([
    [c2, -s2, 0, 8*c2],
    [s2,  c2, 0, 8*s2],
    [0,   0,  1, d2],
    [0,   0,  0, 1]
])

# A3 (with pi/2 + theta3 kept as-is)
A3 = sp.Matrix([
    [c3p, 0,  s3p, 0],
    [s3p, 0, -c3p, 0],
    [0,   1,  0,   0],
    [0,   0,  0,   1]
])

# A4 (using 8 as in your matrix; swap 8 -> d4 if needed)
A4 = sp.Matrix([
    [c4, 0, -s4, 0],
    [s4, 0,  c4, 0],
    [0, -1,  0,  8],
    [0,  0,  0,  1]
])

# A5 (with pi/2 - theta5 kept as-is)
A5 = sp.Matrix([
    [c5m,  0,  s5m, 0],
    [-s5m, 0,  c5m, 0],
    [0,   -1,  0,   0],
    [0,    0,  0,   1]
])

# A6
A6 = sp.Matrix([
    [c6, -s6, 0, 0],
    [s6,  c6, 0, 0],
    [0,    0, 1, d6],
    [0,    0, 0, 1]
])

# Full forward kinematics
T06 = sp.simplify(A1 * A2 * A3 * A4 * A5 * A6)

# Pretty print
sp.pprint(T06)

# LaTeX output
print("\nLaTeX form:\n")
print(sp.latex(T06))
```

```python
import pandas as pd
import numpy as np
import math

# Define the DH table directly as a DataFrame
dh = pd.DataFrame([
    # a, alpha, d, theta, joint_type
    [0,     0,   'd1',   'theta1', 'R'],   # Joint 1: revolute
    [0,   -90,   'd2',   0,        'P'],   # Joint 2: prismatic
    [0,     0,   'd3',   0,        'P'],   # Joint 3: prismatic
], columns=['a', 'alpha', 'd', 'theta', 'type'])

print(dh)
```

```python
def dh_to_A(a, alpha, d, theta):
    alpha = math.radians(alpha)
    theta = math.radians(theta)
    ca, sa = math.cos(alpha), math.sin(alpha)
    ct, st = math.cos(theta), math.sin(theta)

    return np.array([
        [ct, -st*ca,  st*sa, a*ct],
        [st,  ct*ca, -ct*sa, a*st],
        [0,      sa,     ca,   d ],
        [0,       0,      0,   1 ]
    ])

def forward_kinematics(dh, joint_values):
    T = np.eye(4)

    for _, row in dh.iterrows():
        a = float(row['a'])
        alpha = float(row['alpha'])

        # Resolve d and theta depending on joint type
        if row['type'] == 'R':
            theta = float(joint_values[row['theta']])
            d = float(joint_values[row['d']]) if isinstance(row['d'], str) else float(row['d'])
        else:  # prismatic
            d = float(joint_values[row['d']])
            theta = float(joint_values[row['theta']]) if isinstance(row['theta'], str) else float(row['theta'])

        A = dh_to_A(a, alpha, d, theta)
        T = T @ A

    return T
```

```python
joint_values = {
    'theta1': 45,   # degrees
    'd1': 0.1,      # meters
    'd2': 0.3,
    'd3': 0.5
}

T03 = forward_kinematics(dh, joint_values)
print(T03)
```
