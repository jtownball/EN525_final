# Mathematical White Paper

## Introduction

A Selective Compliance Assembly Robot Arm (SCARA) is a type of robot arm that has the RRP configuration, which means it has three joints, two rotary and one prismatic.

## Equations of motion

For the links, $l_1, l_2$, to find the angular velocities we use $\omega = \dot{\theta}\hat{k}$. 
$$\omega_0 = \dot{\theta_0}\hat{k_0}$$
$$\omega_1 = \dot{\theta_1}\hat{k_1}$$
$$\omega_0^2 = \omega_0 + \omega_1 = (\dot{\theta_0} + \dot{\theta_1})\hat{k_0}$$

Because the MTB robot can be treated as a 2R manipulator with prismatic joint, we can directly differentiate the transformation matrices to find the Jacobian by first finding $T_0^2$.

$$
T_0^2 = T_1^0 T_2^1
=\begin{bmatrix}
\cos(\theta_1 + \theta_2) & -\sin(\theta_1 + \theta_2) & 0 & l_1 \cos \theta_1 + l_2 \cos(\theta_1 + \theta_2) \\ 
\sin(\theta_1 + \theta_2) & \cos(\theta_1 + \theta_2) & 0 & l_1 \sin \theta_1 + l_2 \sin(\theta_1 + \theta_2) \\ 
0 & 0 & 1 & 0 \\
0 & 0 & 0 & 1
\end{bmatrix}
$$

$$
\begin{bmatrix}
X \\
Y \\
Z
\end{bmatrix}
=
\begin{bmatrix}
l_1 \cos \theta_1 + l_2 \cos(\theta_1 + \theta_2) \\
l_1 \sin \theta_1 + l_2 \sin(\theta_1 + \theta_2) \\
d_0 + d_3
\end{bmatrix}
$$
Differentiating we get
$$
\begin{bmatrix}
\dot X \\
\dot Y \\
\dot Z
\end{bmatrix}
=
\begin{bmatrix}
-,l_{1}\sin\theta_{1} - l_{2}\sin(\theta_{1}+\theta_{2}) & -,l_{2}\sin(\theta_{1}+\theta_{2}) & 0 \\
l_{1}\cos\theta_{1} + l_{2}\cos(\theta_{1}+\theta_{2}) & l_{2}\cos(\theta_{1}+\theta_{2}) & 0 \\
0 & 0 & 1
\end{bmatrix}
\begin{bmatrix}
\dot \theta_1 \\
\dot \theta_2 \\
\dot d_3
\end{bmatrix}
$$  

Thus, our Jacobian is,
$$
J_D =
\begin{bmatrix}
\frac{\partial X}{\partial \theta_1} & \frac{\partial X}{\partial \theta_2} & \frac{\partial X}{\partial d_3} \\
\frac{\partial Y}{\partial \theta_1} & \frac{\partial Y}{\partial \theta_2} & \frac{\partial Y}{\partial d_3} \\
\frac{\partial Z}{\partial \theta_1} & \frac{\partial Z}{\partial \theta_2} & \frac{\partial Z}{\partial d_3}
\end{bmatrix}
=
\begin{bmatrix}
-,l_1 \sin \theta_1 - l_2 \sin(\theta_1 + \theta_2) & -,l_2 \sin(\theta_1 + \theta_2) & 0 \\
l_1 \cos \theta_1 + l_2 \cos(\theta_1 + \theta_2) & l_2 \cos(\theta_1 + \theta_2) & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

This can be put into standard form as:
$$
\dot{\mathbf{X}} = J_{D} \dot{\boldsymbol{\theta}}
$$

where $\dot{\mathbf{X}} = [\dot X, \dot Y, \dot Z]^T$ and $\dot{\boldsymbol{\theta}} = [\dot \theta_1, \dot \theta_2, \dot d_3]^T$.

## Inertia Tensor
$$
M(\theta) =
\begin{bmatrix}
M_{11} & M_{12} & 0 \\
M_{12} & M_{22} & 0 \\
0 & 0 & m_3
\end{bmatrix}
$$
We will assume a homogenous mass for each ROD, $l_1 \text{, and} l_2$ and we place the center of mass (COM) at the midpoint. $m_3$ is the mass of the end-effector or prismatic joint payload.
$$
I_{1} = \frac{1}{12}\,m_{1} l_{1}^{2}, \qquad
I_{2} = \frac{1}{12}\,m_{2} l_{2}^{2},
\qquad
r_{1} = \frac{l_{1}}{2}, \qquad
r_{2} = \frac{l_{2}}{2}
$$

$$
\begin{aligned}
M_{11} &= I_{1} + I_{2}
+ m_{1} r_{1}^{2}
+ m_{2}\bigl(l_{1}^{2} + r_{2}^{2} + 2\,l_{1} r_{2} \cos\theta_{2}\bigr) \\ 
M_{12} &= I_{2}
+ m_{2}\bigl(r_{2}^{2} + l_{1} r_{2} \cos\theta_{2}\bigr) \\ 
M_{22} &= I_{2} + m_{2} r_{2}^{2}
\end{aligned}
$$

## Potential Energy and Christoffel Symbols
Let gravity be $g$ and potential energy $U(\theta, d_3)$, then
$$
U(\theta, d_3) = g \left[ m_1 r_1 \sin \theta_1 + m_2 (l_1 \sin \theta_1 + r_2 \sin(\theta_1 + \theta_2)) + m_3 (d_0 + d_3) \right]
$$

The Christoffel Symbols for the joint space inertia matrix $M(\theta)$ are,
$$
\Gamma_{ijk}
= \tfrac{1}{2}\biggl(
\frac{\partial M_{ij}}{\partial\theta_k}
+ \frac{\partial M_{ik}}{\partial\theta_j}
- \frac{\partial M_{jk}}{\partial\theta_i}
\biggr)
$$

Focusing on the 3DOF system with $h = m_{2} l_{1} r_{2} \sin\theta_{2}$, the non-zero symbols are,
$$
\Gamma_{112} = \Gamma_{121} = \Gamma_{122} = -h,
\qquad
\Gamma_{211} = +h
$$

All other $\Gamma_{ijk} = 0$, including those involving the third coordinate.
