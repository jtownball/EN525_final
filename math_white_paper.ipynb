## Equations of motion

For the links, $l_1, l_2$, to find the angular velocities we use $\omega = \dot{\theta}\hat{k}$. 
$$\omega_0 = \dot{\theta_0}\hat{k_0}$$
$$\omega_1 = \dot{\theta_1}\hat{k_1}$$
$$\omega_0^2 = \omega_0 + \omega_1 = (\dot{\theta_0} + \dot{\theta_0})\hat{k_0}$$

Because the MTB robot can be treated as a 2R manipulator, we can directly differentiate the transformation matrices to find the Jacobian by first finding $T_0^2$.

$$
T_0^2 = T_1^0 T_2^1
=\begin{bmatrix}
\cos(\theta_1 + \theta_2) & -\sin(\theta_1 + \theta_2) & 0 & l_1 \cos \theta_1 + l_2 \cos(\theta_1 + \theta_2) \\ 
\sin(\theta_1 + \theta_2) & \cos(\theta_1 + \theta_2) & 0 & l_1 \sin \theta_1 + l_2 \sin(\theta_1 + \theta_2) \\ 
0 & 0 & 1 & 0 \\
0 & 0 & 0 & 1
\end{bmatrix}
$$

$$
\begin{bmatrix}
X \\
Y
\end{bmatrix}
=
\begin{bmatrix}
l_1 \cos \theta_1 + l_2 \cos(\theta_1 + \theta_2) \\
l_1 \sin \theta_1 + l_2 \sin(\theta_1 + \theta_2)
\end{bmatrix}
$$
Differentiating we get
$$
\begin{bmatrix}
\dot X \\
\dot Y
\end{bmatrix}
=
\begin{bmatrix}
-,l_{1}\sin\theta_{1} - l_{2}\sin(\theta_{1}+\theta_{2}) & -,l_{2}\sin(\theta_{1}+\theta_{2}) \\
l_{1}\cos\theta_{1} + l_{2}\cos(\theta_{1}+\theta_{2}) & l_{2}\cos(\theta_{1}+\theta_{2})
\end{bmatrix}
$$  

Thus, our Jacobian is,
$$
J_D =
\begin{bmatrix}
\frac{\partial X}{\partial \theta_1} & \frac{\partial X}{\partial \theta_2} \\
\frac{\partial Y}{\partial \theta_1} & \frac{\partial Y}{\partial \theta_2}
\end{bmatrix}
=
\begin{bmatrix}
-,l_1 \sin \theta_1 - l_2 \sin(\theta_1 + \theta_2) & -,l_2 \sin(\theta_1 + \theta_2) \\
l_1 \cos \theta_1 + l_2 \cos(\theta_1 + \theta_2) & l_2 \cos(\theta_1 + \theta_2)
\end{bmatrix}
$$

This can be put into standard form as:
$$
\dot{\mathbf{X}} = J_{D},\dot{\boldsymbol{\theta}}
$$


## Inertia Tensor
$$
M(\theta) =
\begin{bmatrix}
M_{11} & M_{12} \\
M_{12} & M_{22}
\end{bmatrix}
$$
We will assume a homogenous mass for each ROD, $l_1 \text{, and} l_2$ and we place the center of mass (COM) at the midpoint.
$$
I_{1} = \frac{1}{12},m_{1} l_{1}^{2}, \qquad
I_{2} = \frac{1}{12},m_{2} l_{2}^{2},
\qquad
r_{1} = \frac{l_{1}}{2}, \qquad
r_{2} = \frac{l_{2}}{2}.
$$

$$
\begin{aligned}
M_{11} &= I_{1} + I_{2}
+ m_{1} r_{1}^{2}
+ m_{2}\bigl(l_{1}^{2} + r_{2}^{2} + 2,l_{1} r_{2} \cos\theta_{2}\bigr) \\ 
M_{12} &= I_{2}
+ m_{2}\bigl(r_{2}^{2} + l_{1} r_{2} \cos\theta_{2}\bigr) \\ 
M_{22} &= I_{2} + m_{2} r_{2}^{2}.
\end{aligned}
$$

## Potential Energy and Christoffel Symobls
Let gravity be $g$ and potential energy $U(\theta)$, then
$$
U(\theta) = g\bigl[m_{1} r_{1}\sin\theta_{1}
+ m_{2}\bigl(l_{1}\sin\theta_{1} + r_{2}\sin(\theta_{1}+\theta_{2})\bigr)\bigr]
$$
The Christoffel Symbols for the joint space inertia matrix $M(\theta)$ are,
$$
\Gamma_{ijk}
= \tfrac{1}{2}\biggl(
\frac{\partial M_{ij}}{\partial\theta_k}

\frac{\partial M_{ik}}{\partial\theta_j}
\frac{\partial M_{jk}}{\partial\theta_i}
\biggr)
$$

Focusing on the 2R planner portion with $h = m_{2} l_{1} r_{2} \sin\theta_{2}$, the non-zero symbols are,
$$
\Gamma_{112} = \Gamma_{121} = \Gamma_{122} = -h
\qquad
\Gamma_{211} = +h
$$

All other $\Gamma_{ijk} = 0$
